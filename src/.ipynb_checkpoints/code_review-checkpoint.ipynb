## みんなの手元でやってほしいこと
### anaconda で仮想環境を作る
コマンドプロンプトで以下を入力。
```
conda create -n review python=3.7 anaconda
```
すると
```
Proceed ([y]/n)?
```
と聞かれるので、yを入力してください。
しばらくすると、仮想環境が出来上がります。その後、コマンドプロンプトで以下を入力。
```
conda activate review
```
### 必要ライブラリのインストール
このコードは以下の３つのライブラリを使用しているため、インストールする必要があります。
* opencv

コマンドプロンプトで以下を入力してください。

Proceed ([y]/n)?

````python
from conduct_csc import*;
from visualize import*;
from load_dataset import*;
from scipy import sparse;
````

````python
def make_option():
    opt = {};
    # 訓練画像の枚数
    opt['train_amount'] = 100;
    # テスト画像の枚数
    opt['test_amount'] = 100;
    # フィルタのサイズ
    opt['d_size'] = [(5, 5, 6), (5, 5, 6, 16)];
    # スパースの尺度
    opt['lmbda'] = [0.5, 0.025];
    # 繰り返し回数
    opt['Iter'] = [200, 200];
    
    return opt;
````

````python
# 多層CSCの実行
opt = make_option();
train_data, train_label = load_mnist_train(opt["train_amount"]);
d0, D0, coef0 = nn_csc(input_ = train_data, d_size = opt['d_size'][0], lmbda = opt['lmbda'][0], Iter = opt['Iter'][0], visualize = True);
d1, D1, coef1 = nn_csc(input_ = coef0.squeeze().transpose(0,1,3,2) , d_size = opt['d_size'][1], lmbda = opt['lmbda'][1], Iter = opt['Iter'][1], visualize = False);
# output(opt, D0, coef0, D1, coef1);
````
